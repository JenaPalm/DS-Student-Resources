```python
import pandas as pd
```

```python
# max out those columns and rows
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
# pull the latest down from https://marketplace.fedramp.gov/#!/products?sort=productName
fedRAMP = pd.read_csv("../Data/fedramp-2023-03-02.csv")
```

```python
fedRAMP.info()
```

```python
#hmm definitely want to rename these columns
fedRAMP.columns
```

```python
fedRAMP.tail()
```

```python
# rename columns to remove spaces and convert to lowercase
df = fedRAMP.rename(columns={
    ' Cloud Service Provider': 'provider',
    ' Cloud Service Offering': 'offering',
    ' Type': 'type',
    ' Service Model': 'model',
    ' Leveraged A T O': 'ato',
    ' Authorization Date': 'authorization_date',
    ' Announcement Date': 'announcement_date',
    ' Agency': 'agency',
    ' Subagency': 'subagency',
    ' Independent Assessor': 'assessor',
    ' Package Id': 'package_id'
})
```

```python
df.head()
```

```python
# filter rows based on a condition
eighteenf = df[df['provider'] == '18F']
```

```python
eighteenf
```

```python
# group data by a provider column and count package_id column
grouped = df.groupby('provider')['package_id'].count()
```

```python
grouped
```

```python
#group data by provider column and calculate the count  column, then sort by count
grouped = df.groupby('provider')['package_id'].count().sort_values(ascending=False)
print(grouped)
```

```python
#group data by provider column and calculate the count  column, then sort by count
grouped = df.groupby('agency')['package_id'].count().sort_values(ascending=False)
print(grouped)
```
