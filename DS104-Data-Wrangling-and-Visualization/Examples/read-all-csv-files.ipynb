```python
#name necessary variables and print out data path
path=os.getcwd()
data='/Data'
data_path=path+data

print(data_path)
```

```python
#save all datasets to a callable list of pandas dataframes
csv_files = []
for file in os.listdir(data_path):
    if file.endswith(".csv"):
        csv_files.append(file)
        
df = {}

for file in csv_files:
    try:
        df[file] = pd.read_csv(data_path+'/'+file)
    except UnicodeDecodeError:
        df[file] = pd.read_csv(data_path+'/'+file, encoding="ISO-8859-1") #if utf-8 encoding error
        
for k in csv_files:

    #call dataframe
    dataframe = df[k]
    
print(csv_files)
```
