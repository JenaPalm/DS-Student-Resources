# Hotel example in R

This notebook uses hotel data to demonstrate a variety of plot options in R. 

```r
# Install packages
install.packages("tidyverse")
```

```r
# Load in Libraries
library("tidyverse")
```

```r
# Load in data
hotels = read_csv("../../Data/hotels.csv")
```

```r
# View data
head(hotels)
```

# Bar charts

```r
ggplot(data=hotels, aes(x=hotel)) + geom_bar()
```

```r
ggplot(data=hotels, aes(x=arrival_date_year)) + geom_bar()
```

```r
ggplot(data=hotels, aes(x=deposit_type)) + geom_bar()
```

```r
ggplot(data=hotels, aes(x=customer_type)) + geom_bar()
```

```r
ggplot(data=hotels, aes(x=reservation_status)) + geom_bar()
```

# Scatter plots

```r
ggplot(data=hotels, aes(x=adults, y=children)) + geom_point()
```

```r
ggplot(data=hotels, aes(x=adults, y=babies)) + geom_point()
```

```r
ggplot(data=hotels, aes(x=stays_in_week_nights, y=children)) + geom_point()
```

# Line chart

```r
ggplot(data=hotels, aes(x=reservation_status_date, y=booking_changes)) + geom_line() 
```
