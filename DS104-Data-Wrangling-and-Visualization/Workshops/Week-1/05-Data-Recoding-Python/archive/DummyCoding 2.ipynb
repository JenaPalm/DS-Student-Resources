# Import Packages

```python
import pandas as pd
```

```python
glassdoor = pd.read_csv('../../Data/glassdoor.csv')
glassdoor.head()
```

# Dummy Coding the Industry

```python
IndustryDummy = pd.get_dummies(glassdoor['Industry'], drop_first=True)
IndustryDummy.head()
```

```python
glassdoor1 = pd.concat([glassdoor, IndustryDummy], axis=1)
glassdoor1.head()
```

# Continuous to Categorical

```python
def salary (series):
    if series < 30000:
        return 0
    if series >= 60000:
        return 1

glassdoor1['MaxSalaryR'] = glassdoor1['Max_Salary'].apply(salary)
```
