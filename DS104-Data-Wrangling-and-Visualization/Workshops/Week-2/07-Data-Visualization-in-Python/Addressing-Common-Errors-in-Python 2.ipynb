```python
import pandas as pd
import seaborn as sns
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

```python
flowers = pd.read_csv("flowers.csv")
flowers.head()
```

```python
flowerShipment = pd.read_csv("flowerShipment.csv")
flowerShipment.head()
```

```python
flowerShipment.info()
```

```python
flowerShipment['Date'] = flowerShipment['Date'].astype('datetime64[ns]')
```

```python
flowerShipment.info()
```

```python
flowerShipment2 = flowerShipment.drop_duplicates(subset='Date', keep='first')
```

```python
flowerShipment.sort_values(by=['Date'], inplace=True)
```

## Histograms with pandas

```python
flowers['Days to grow'].hist()
```

## Single Histogram with seaborn

```python
sns.distplot(flowers['Days to grow'])
```

## Multiple Histograms with seaborn

```python
sns.pairplot(flowers)
```

## Histograms with matplotlib

```python
num_bins = 12
n, bins, patches = plt.hist(flowers['Days to grow'], num_bins, facecolor='green', alpha=.7)
plt.xlabel('Days to Grow')
plt.ylabel('Frequency')
plt.title('Histogram of the Number of Days to Grow')
plt.show()
```

## Bar Charts

```python
BarChart = flowers['type of flower'].value_counts().plot('bar')
BarChart.set_title("Frequency of Flower Types")
BarChart.set_xlabel("Flower Types")
BarChart.set_ylabel("Frequency")
```

## Scatter Plots

```python
ScatterPlot = flowers.plot.scatter(x='Cost to grow', y='Cost of force', c='Work force', cmap='coolwarm')
ScatterPlot.set_title("Cost to Grow Flowers by Cost of Force")
ScatterPlot.set_xlabel("Cost to Grow Flowers")
ScatterPlot.set_ylabel("Cost of Force")
```

```python
ScatterPlot = flowers.plot.scatter(x='Cost to grow', y='Cost of force', s=flowers['Work force']*4)
ScatterPlot.set_title("Cost to Grow Flowers by Cost of Force")
ScatterPlot.set_xlabel("Cost to Grow Flowers")
ScatterPlot.set_ylabel("Cost of Force")
```
