```r
# Load in Libraries
library("ggplot2") 
library("lattice")
library("IDPmisc")
```

```r
flowers = read.csv("../../Data/flowers.csv")
```

```r
dim(flowers)
```

```r
head(flowers)
```

```r
unique(flowers$type.of.flower)
```

```r
# Histogram in R
ggplot(flowers, aes(x = Days.to.grow)) + geom_histogram(bins=50) # Bar Chart with ggplot
```

```r
# Bar chart with lattice
barchart(flowers$type.of.flower, main="Frequency of Flower Type", ylab="Flower Type", xlab="Frequency")
```

```r
ggplot(flowers, aes(type.of.flower))+ geom_bar() +
ggtitle("Frequency of Flower Types") + xlab("Flower Types") + ylab("Frequency")
```

```r
# Scatter Plot in R
ggplot(flowers, aes(x = Cost.to.grow, y = Cost.of.force)) + geom_point()
```

```r
flowerShipment = read.csv("../../Data/flowerShipment.csv")
```

```r
dim(flowerShipment)
```

```r
head(flowerShipment)
```

```r
# Line Charts in R
ggplot(flowerShipment, aes(as.Date(Date, "%m/%d/%Y"), Cost)) + geom_line() +
xlab("Date of Flower Shipment") +
ylab("Cost of Flowers") +
ggtitle("Cost of Flowers Over Time")
```

```r
unique(flowerShipment$Color)
```

```r
ggplot(flowerShipment, aes(x=Color, y=Cost, fill=Color)) + 
geom_bar(stat = "summary", fun=mean) + 
theme(axis.text.x=element_text(angle=90, hjust=1))
```
